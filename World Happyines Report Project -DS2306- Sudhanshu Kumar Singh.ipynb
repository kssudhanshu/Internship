The dependent variable is happiness score, and the independent variables are economy, family, life expectancy,freedom, generosity, trust, and dystopia residual.

```python
#X = df['attend'].values.reshape(-1,1)
#y = df['temp'].values.reshape(-1,1)

X = data.drop(['Happiness_Score', 'Happiness_Rank', 'Country','Region'], axis=1)
#X = df_new.drop(['Happiness.Score', 'Happiness.Rank'], axis=1)
y = data['Happiness_Score']
X.head()

```

```python
y.head()
```

```python
# i will split 80% data for training and 20% data for testing 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# importing leanier regrassion 
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train, y_train) #training the algorithm

#regressor = LinearRegression()  
#regressor.fit(X_train, y_train) #training the algorithm
```

The linear regression model basically finds the best value for the intercept and slope, which results in a line that best fits the data.

```python
#To retrieve the intercept:
print(lm.intercept_)#For retrieving the slope:
print(lm.coef_)
```

```python

print('Coefficients: \n', lm.coef_)
lm.coef_

#Note - one unit of change in x, y change to 0.999722
```

```python
predictions = lm.predict( X_test)
predictions
```

```python
#
lm.score(X_test, y_test)
```

```python
#comparing the Actual and pridicted value of x 
data = pd.DataFrame({'Actual': y_test, 'Predicted': predictions})
data.head()
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('X_Test')
plt.ylabel('Predicted Y')

#scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(y_test, predictions, color='skyblue', edgecolor='purple', alpha=0.7, marker='o')

# Adding title and labels
plt.title('Comparison of Actual vs. Predicted', fontsize=16)
plt.xlabel('Actual Y', fontsize=14)
plt.ylabel('Predicted Y', fontsize=14)

# Add a legend
plt.legend(['Data Points'], loc='best', fontsize=12)

```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
#The sign of each coefficient indicates the direction of the relationship between a predictor variable and the response variable.
#A positive sign indicates that as the predictor variable increases, the response variable also increases.
#A negative sign indicates that as the predictor variable increases, the response variable decreases.


coeffecients = pd.DataFrame(lm.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```
